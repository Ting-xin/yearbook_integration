```python
import os
import json


with open('./江苏省分地区城镇常住居民家庭基本情况.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

print(data)
```

```python
result = []
for i in data:
    temp = i
    temp['space'] = i['index'][0]
    temp['index'] = ' '.join(list(set(temp['index'])))
    temp['unit'] = ' '.join(list(set(temp['unit'])))
    result.append(temp)

with open('temp_data.json', 'w', encoding='utf-8') as file:
    json.dump(result, file, ensure_ascii=False, indent=4)
    
```
